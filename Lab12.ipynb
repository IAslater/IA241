```python
import pandas 

df = pd.read_excel("s3://slater-ia241-spring2023/house_price.xls")
df.head(10)
```

#3.1


```python
df["unit_price"] = df['price']/df['area']
df.head(10)
```

```python
df["house_type"].value_counts()
```

```python
print("The average price for homes with more than 2 bathrooms is ${:.2f}".format(df.loc[df["bathroom"] > 2]["price"].mean()))          
```

#3.4


```python
print("mean price is ${:.2f}".format(df['unit_price'].mean()))
```

```python
print("median price is ${:.2f}".format(df['unit_price'].median()))
```

```python
from scipy import stats
```

```python
model = stats.linregress(df['area'], df['price'])
```

```python
print("slope is {:.2f}".format(model.slope))
print("intercept is {:.2f}".format(model.intercept))
print("p-value is {:.2f}".format(model.pvalue))
print("R square is {:.2f}".format(model.rvalue ** 2))
```

```python
corr = df['price'].corr(df['lot_size'])
print('The correlation between price and lot size is {:.2f}.'.format(corr))
```
